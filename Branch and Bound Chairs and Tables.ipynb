```python
from pulp import *

# We intentially made the documentation sparse on this one so we could copy the code for branch and bound

#First Iteration of the solve
prob = LpProblem("Branch and Bound Sample",LpMaximize)
X1=LpVariable("X1 Table",0,None,LpContinuous)
X2=LpVariable("X2 Chair",0,None, LpContinuous)
prob += 8*X1 + 5*X2,"Total Profit"
prob += X1 + X2 <= 6, "Labor Constraint"
prob += 9*X1 + 5*X2 <= 45, "Wood Constraint"
prob += X1 >= 0, "LB for X1"
prob += X1 <= 999, "UB for X1"
prob += X2 >= 0, "LB for X2"
prob += X2 <=999, "UB for X2"
prob.writeLP("TableChair.lp")
prob.solve()
print("Iteration 1")
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Total Profit = ", value(prob.objective))
print("--------------------------")

#2nd Iteration of the solve
prob = LpProblem("Branch and Bound Sample",LpMaximize)
X1=LpVariable("X1 Table",0,None,LpContinuous)
X2=LpVariable("X2 Chair",0,None, LpContinuous)
prob += 8*X1 + 5*X2,"Total Profit"
prob += X1 + X2 <= 6, "Labor Constraint"
prob += 9*X1 + 5*X2 <= 45, "Wood Constraint"
prob += X1 >= 4, "LB for X1"
prob += X1 <= 999, "UB for X1"
prob += X2 >= 0, "LB for X2"
prob += X2 <=999, "UB for X2"
prob.writeLP("TableChair.lp")
prob.solve()
print("Iteration 2")
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Total Profit = ", value(prob.objective))
print("--------------------------")

#3rd Iteration of the solve
prob = LpProblem("Branch and Bound Sample",LpMaximize)
X1=LpVariable("X1 Table",0,None,LpContinuous)
X2=LpVariable("X2 Chair",0,None, LpContinuous)
prob += 8*X1 + 5*X2,"Total Profit"
prob += X1 + X2 <= 6, "Labor Constraint"
prob += 9*X1 + 5*X2 <= 45, "Wood Constraint"
prob += X1 >= 0, "LB for X1"
prob += X1 <= 3, "UB for X1"
prob += X2 >= 0, "LB for X2"
prob += X2 <=999, "UB for X2"
prob.writeLP("TableChair.lp")
prob.solve()
print("Iteration 3")
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Total Profit = ", value(prob.objective))
print("--------------------------")

#4th Iteration of the solve
prob = LpProblem("Branch and Bound Sample",LpMaximize)
X1=LpVariable("X1 Table",0,None,LpContinuous)
X2=LpVariable("X2 Chair",0,None, LpContinuous)
prob += 8*X1 + 5*X2,"Total Profit"
prob += X1 + X2 <= 6, "Labor Constraint"
prob += 9*X1 + 5*X2 <= 45, "Wood Constraint"
prob += X1 >= 4, "LB for X1"
prob += X1 <= 999, "UB for X1"
prob += X2 >= 2, "LB for X2"
prob += X2 <=999, "UB for X2"
prob.writeLP("TableChair.lp")
prob.solve()
print("Iteration 4")
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Total Profit = ", value(prob.objective))
print("--------------------------")

#5th Iteration of the solve
prob = LpProblem("Branch and Bound Sample",LpMaximize)
X1=LpVariable("X1 Table",0,None,LpContinuous)
X2=LpVariable("X2 Chair",0,None, LpContinuous)
prob += 8*X1 + 5*X2,"Total Profit"
prob += X1 + X2 <= 6, "Labor Constraint"
prob += 9*X1 + 5*X2 <= 45, "Wood Constraint"
prob += X1 >= 4, "LB for X1"
prob += X1 <= 999, "UB for X1"
prob += X2 >= 0, "LB for X2"
prob += X2 <=1, "UB for X2"
prob.writeLP("TableChair.lp")
prob.solve()
print("Iteration 5")
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Total Profit = ", value(prob.objective))
print("--------------------------")

#6th Iteration of the solve
prob = LpProblem("Branch and Bound Sample",LpMaximize)
X1=LpVariable("X1 Table",0,None,LpContinuous)
X2=LpVariable("X2 Chair",0,None, LpContinuous)
prob += 8*X1 + 5*X2,"Total Profit"
prob += X1 + X2 <= 6, "Labor Constraint"
prob += 9*X1 + 5*X2 <= 45, "Wood Constraint"
prob += X1 >= 5, "LB for X1"
prob += X1 <= 999, "UB for X1"
prob += X2 >= 0, "LB for X2"
prob += X2 <=1, "UB for X2"
prob.writeLP("TableChair.lp")
prob.solve()
print("Iteration 6")
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Total Profit = ", value(prob.objective))
print("--------------------------")

#7th Iteration of the solve
prob = LpProblem("Branch and Bound Sample",LpMaximize)
X1=LpVariable("X1 Table",0,None,LpContinuous)
X2=LpVariable("X2 Chair",0,None, LpContinuous)
prob += 8*X1 + 5*X2,"Total Profit"
prob += X1 + X2 <= 6, "Labor Constraint"
prob += 9*X1 + 5*X2 <= 45, "Wood Constraint"
prob += X1 >= 0, "LB for X1"
prob += X1 <= 4, "UB for X1"
prob += X2 >= 0, "LB for X2"
prob += X2 <=1, "UB for X2"
prob.writeLP("TableChair.lp")
prob.solve()
print("Iteration 7")
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Total Profit = ", value(prob.objective))
print("--------------------------")
```
